```r
quants = quantile(prostate$PSA, c(0, 0.1, 0.5, 0.9))
quantile(prostate$PSA, 0.1)
```

```r
PSALevels <- c('low', 'med', 'high', 'extreme')
prostate$classPSA <- findInterval(prostate$PSA, quants)
prostate$classPSA <- PSALevels[prostate$classPSA]
prostate
```

```r

table(prostate$classPSA, prostate$Gleason)
```

```r
boxplot(PSA ~ Gleason, prostate)
boxplot(vol ~ Gleason, prostate)
boxplot(weight ~ Gleason, prostate)
```

```r
PSAnorm <- tapply(prostate$PSA, prostate$Gleason, function (x) ((x-mean(x))/sd(x)))
qqplot(PSAnorm[[1]], PSAnorm[[2]])
abline(0, 1)
qqplot(PSAnorm[[2]], PSAnorm[[3]])
                  abline(0, 1)
qqplot(PSAnorm[[1]], PSAnorm[[3]])
abline(0, 1)
```

```r
qqnorm(PSAnorm[[1]])
qqnorm(PSAnorm[[2]])
qqnorm(PSAnorm[[3]])
```
